```python
import h5py

# Open the HDF5 file
with h5py.File('trained_model.h5', 'r') as h5_file:
    # List all groups (datasets, layers, etc.)
    def print_structure(name, obj):
        print(f"{name}: {obj}")

    # Recursively print the structure of the file
    h5_file.visititems(print_structure)
```
